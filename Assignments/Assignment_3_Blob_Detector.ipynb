```python
import cv2
import numpy as np

image = cv2.imread("Images/SHAPES.png", 0)
cv2.imshow('Original Image',image)
cv2.waitKey(0)


params = cv2.SimpleBlobDetector_Params()
#Setting parameters false because with the default it was not detecting all shapes

params.filterByArea = False

params.filterByCircularity = False

params.filterByConvexity = False

params.filterByInertia = False

detector = cv2.SimpleBlobDetector_create(params)
 
keypoints = detector.detect(image)

blank = np.zeros((1,1)) 
blobs = cv2.drawKeypoints(image, keypoints, blank, (255,0,0),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

number_of_blobs = len(keypoints)
text = "Total No. of Blobs identified: " + str(len(keypoints))
cv2.putText(blobs, text, (400, 600), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 0, 0), 2)

cv2.imshow("All Shapes using default parameters", blobs)
cv2.waitKey(0)


#parameters to find circles
param = cv2.SimpleBlobDetector_Params()

param.filterByArea = False
#param.minArea = 0

param.filterByCircularity = True 
param.minCircularity = 0.892

param.filterByConvexity = True
param.minConvexity = 0.3 

param.filterByInertia = True
param.minInertiaRatio = 0.05

detector = cv2.SimpleBlobDetector_create(param)    

keypoints = detector.detect(image)

blank = np.zeros((1,1)) 
blobs = cv2.drawKeypoints(image, keypoints, blank, (0,255,0),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

number_of_blobs = len(keypoints)
text = "No. of Circles: " + str(len(keypoints))
cv2.putText(blobs, text, (420, 600), cv2.FONT_HERSHEY_DUPLEX, 1, ( 0,255,0),2)

cv2.imshow("Showing Circles", blobs)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
